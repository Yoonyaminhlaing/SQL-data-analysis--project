```python
import sqlite3
import pandas as pd
```

```python
df = pd.read_csv("/content/sample_data/encoded-Sample - Superstore - Modified -.csv")
```

```python
conn = sqlite3.connect('Sale_database')
c = conn.cursor()
```

```python
df.to_sql('sales', conn, if_exists='replace', index = False)
```

```python
c.execute('''
  SELECT State, SUM(Profit) FROM sales
  GROUP BY State
''')
for row in c.fetchall():
    print (row)
```

```python
c.execute('''
  SELECT Region, SUM(Profit) FROM sales
  GROUP BY Region
''')
for row in c.fetchall():
    print (row)
```

```python
c.execute('''
  SELECT Category, SUM(Profit) FROM sales
  GROUP BY Category
''')
for row in c.fetchall():
    print (row)
```

```python
c.execute('''
  SELECT Subcategory, SUM(Profit) FROM sales
  GROUP BY Subcategory
''')
for row in c.fetchall():
    print (row)
```

```python
c.execute('''
  SELECT ProductName, SUM(Profit) AS total_profit FROM sales
  GROUP BY ProductName
  ORDER BY total_profit DESC
  LIMIT 5
''')
for row in c.fetchall():
    print (row)
```

```python
c.execute('''
  SELECT ProductName, SUM(Profit) AS total_profit FROM sales
  GROUP BY ProductName
  ORDER BY total_profit Asc
  LIMIT 5
''')
for row in c.fetchall():
    print (row)
```

```python
#Find the maximum discount applied to any product.
c.execute('''
   SELECT ProductName, MAX(Discount) FROM Sales
   GROUP BY ProductName
   LIMIT 10
''')
for row in c.fetchall():
    print (row)
```

```python
#- Find the total sales for each 'Category' and 'SubCategory'.
c.execute('''
    SELECT Category, sum(sales) FROM Sales
    GROUP BY Category
 ''')
for row in c.fetchall():
    print (row)
```

```python
#- Find the total sales for each 'Category' and 'SubCategory'.
c.execute('''
    SELECT SubCategory, sum(sales) FROM Sales
    GROUP BY SubCategory
 ''')
for row in c.fetchall():
    print (row)
```

```python
#Calculate the total quantity sold for each 'ProductID'.
c.execute('''
   SELECT ProductID, SUM(Quantity) FROM Sales
   Group By ProductID
 ''')
for row in c.fetchall():
    print (row)
```

```python
#Retrieve all records where the 'Sales' are greater than the average sales
c.execute('''
   SELECT * FROM Sales
   WHERE sales > (SELECT AVG(sales) FROM Sales)
''')
for row in c.fetchall():
    print (row)
```

```python
#Find the 'ProductID' with the highest sales.
c.execute('''
    SELECT ProductID FROM Sales
    WHERE Sales = (SELECT MAX(Sales) FROM Sales)
''')
for row in c.fetchall():
    print (row)
```

```python
#Create a new column called 'Revenue' which is the product of 'Sales' and 'Quantity'.
c.execute('''
   ALTER TABLE Sales
   ADD revenue DECIMAL(10,2)
''')

c.execute('''
    UPDATE Sales
    SET revenue = Sales * Quantity
 ''')
```

```python
c.execute('''
  SELECT revenue FROM Sales
''')
for row in c.fetchall():
    print (row)
```

```python
#Calculate the net profit, considering the discount, by subtracting 'Discount' from 'Profit'.
c.execute('''
    ALTER TABLE Sales
    ADD net_profit DECIMAL(10,2)
 ''')
```

```python
c.execute('''
    UPDATE Sales
    SET net_profit = Profit - Discount
''')
```

```python
c.execute('''
SELECT ProductID, net_profit FROM Sales
''')
for row in c.fetchall():
    print (row)
```

```python
#Retrieve records for products with a discount greater than 0.2.
c.execute('''
    SELECT ProductName, Discount FROM Sales
    WHERE Discount > 0.2
''')
for row in c.fetchall():
    print (row)
```

```python
#Find the records where 'Category' is 'Office Supplies' and 'Quantity' is greater than 5.
c.execute('''
  SELECT * FROM Sales
  WHERE Category = "Office Supplies" AND Quantity > 5
 ''')
for row in c.fetchall():
    print (row)
```

```python
#- Find records where 'Profit' is NULL
c.execute('''
  SELECT * FROM SALES
  WHERE Profit IS NULL
''')
```

```python
#Replace NULL values in the 'Discount' column with 0
c.execute('''
  UPDATE Sales
  SET Discount = 0
  WHERE Discount IS NULL
''')
```
